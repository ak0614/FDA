```python
'''1.) Implement Logistic Regression 
   a.) Import Libraries and Load Dataset
   b.) Analyze the Dataset using Matplotlib
   c.) Data Wrangling 
   d.) Train & fIt Logistic Regression Model
   e.) Check Accuracy Score 
   f.) Predict Values with External CSV File'''
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
Data=pd.read_csv("Logistic_regression.csv")
plt.scatter(Data.age,Data.bought_insurance)
plt.show()
Data=Data.replace(to_replace="NAN",value=0)
reg=LogisticRegression()
reg.fit(Data[["age"]],Data.bought_insurance)
n=int(input("Enter a value(age) to predict the value for this model:"))
print("The accuracy of the predicted model is:",reg.score(Data[["age"]],Data.bought_insurance)*100)
print("Prediction of Data and the value for the age",n," is:",reg.predict([[n]]))
```

```python
'''2.) Implement Logistic Regression 
   a.) Import Libraries and Load Dataset
   b.) Analyze the Dataset using Matplotlib
   c.) Data Wrangling 
   d.) Train & fIt Logistic Regression Model
   e.) Check Accuracy Score 
   f.) Predict Values with External CSV File'''
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
Data=pd.read_csv("CarPrice_Assignment.csv")
Data=Data.replace(to_replace="NaN",value=0)
plt.scatter(Data.CarName,Data.price)
plt.show()
reg=linear_model.LinearRegression()
reg.fit(Data[['horsepower','peakrpm']],Data.price)
i=int(input("Enter a HORSE-POWER for prediction:"))
j=int(input("Enter the PEAK-RPM for prediction:"))
print("The predicted Data is:",reg.predict([[i,j]]))
print("The Accuracy of this model is:",reg.score(Data[['horsepower','peakrpm']],Data.price)*100)
```
