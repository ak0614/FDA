1. Draw a pie chart for following data
a.) start with grocery as 0 degree
b.) Explode with shadow for Saving and Automobile
c.) All Labels incorporate
     Grocery- 21%
     Automobile - 22%
     Utitlity Bills - 12%
     Dining out - 8%
     Healthcare-6%
     Saving -23%
     Other Expenses- 8%

```python
import pandas as pd
import matplotlib.pyplot as plt
data=[21,22,12,8,6,23,8]
mylabels=['Grocery-21%','Automobile-22%','Utitlity Bills-12%','Dining out-8%','Healthcare-6%','Saving-23%','Other Expenses-8%']
myexplode=[0,0.1,0,0,0,0.2,0]
plt.pie(data,labels=mylabels,startangle=0,explode=myexplode,shadow=True)
plt.show()
```

2. Draw a Multiple Line graph for following data with label , x-axis and Y-axis

3. Draw a Bar Chart for following data 
a.) Width of the bar is 0.5
b.) Each bar must be a different color 
c.) Draw horizontal bar with height =0.2
D.) x axis and y axis label should present

```python
y=[12,15,15,26,60]
x=['Goat','Horse','Cow','Camel','Elephant']
mycolor=['red','green','b','c','hotpink']
plt.bar(x,y,width=0.5,color=mycolor)
plt.xlabel('Age(in years)')
plt.ylabel('Name of animals')
plt.show()
plt.barh(x,y,height=0.2,color=mycolor)
plt.xlabel('Age(in years)')
plt.ylabel('Name of animals')
plt.show()
```

0.50 - 0.99		    4

```python
import matplotlib.pyplot as plt
import numpy as np
frequency=np.array([0.7,0.67,0.83,0.53])
plt.hist(frequency,bins=[0.50,1.00,1.50,2.00,2.50,3.00,3.50,4.00],color='red')
frequency1=[1.2,1.32,1.44,1.24,1.13,1.29,1.38]
frequency2=[1.99,1.59,1.93]
frequency3=[2.3,2.1,2.34]
frequency4=[3.36,3.27]
frequency5=[3.69]
plt.hist(frequency1,bins=[0.50,1.00,1.50,2.00,2.50,3.00,3.50,4.00],color='blue')
plt.hist(frequency2,bins=[0.50,1.00,1.50,2.00,2.50,3.00,3.50,4.00],color='green')
plt.hist(frequency3,bins=[0.50,1.00,1.50,2.00,2.50,3.00,3.50,4.00],color='cyan')
plt.hist(frequency4,bins=[0.50,1.00,1.50,2.00,2.50,3.00,3.50,4.00],color='pink')
plt.hist(frequency5,bins=[0.50,1.00,1.50,2.00,2.50,3.00,3.50,4.00],color='black')
plt.xlabel('Price(dollar)')
plt.ylabel('Frequency')
plt.title('Price of Cards')
plt.show()
```

5. Create a grapgh for square of numbers (0-4)

```python
import matplotlib.pyplot as plt
plt.axis([0,5,0,20])
plt.plot([0,1,2,3,4],[0,1,4,9,16],'r.-')
plt.xlabel('Numbers',color='g',fontsize=15)
plt.ylabel('Square of the numbers',color='c',fontsize=15)
plt.text(0,0.5,'Zero')
plt.text(1,1.5,'One')        
plt.text(2,4.5,'Two')
plt.text(3,9.5,'Three')
plt.text(4,16.5,'Four')
plt.text(1.1,12,'$y = x^2$',fontsize=20,bbox={'facecolor':'yellow','alpha':0.5})
plt.grid(True)
plt.legend(['Squares'],loc=7)
plt.title('Graph of Square of Numbers',fontsize=20)
plt.show()
```
